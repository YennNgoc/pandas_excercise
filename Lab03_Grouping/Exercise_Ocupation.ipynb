### Step 4. Discover what is the mean age per occupation

```python
mean_age = user.groupby(['occupation']).mean()
mean_age['age']
```

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least

```python

ratio_1 = user.groupby(['occupation']).count()
user_male = user.loc[user['gender'] == 'M']
ratio = user_male.groupby(['occupation']).count()
a = ratio['gender']
b = ratio_1['gender']
c = pd.merge(a,b,'inner',on='occupation')
c['new']= c['gender_x']/c['gender_y']*100
c.sort_values('new', ascending=False)
```

### Step 6. For each occupation, calculate the minimum and maximum ages

```python
user.head()
max_age = user.groupby(['occupation']).max()
max_age['age']
min_age = user.groupby(['occupation']).min()
min_age['age']
ages_min_max = pd.merge(max_age['age'],min_age['age'],'inner',on='occupation')
ages_min_max
```

### Step 7. For each combination of occupation and gender, calculate the mean age

```python
user.head()
user2 = user.groupby(['occupation','gender'], as_index=False)['age'].mean()
user2
```

### Step 8.  For each occupation present the percentage of women and men

```python
user.head()
ratio_1 = user.groupby(['occupation'])['user_id'].count()
ratio = user.groupby(['occupation','gender'], as_index=False)['user_id'].count()
c = pd.merge(ratio,ratio_1,'inner',on='occupation')
c
c['new']= c['user_id_x']/c['user_id_y']*100
c1 = pd.pivot_table(c,values='new',index='occupation', columns='gender')
c1
```
